```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
```

```python
def show_company_results(ticker):

    # Load metric results
    results = pd.read_csv("evaluation_results.csv")
    company_results = results[results["ticker"] == ticker]
    
    if company_results.empty:
        print(f"No results found for {ticker}")
        return

    print(f"\nEvaluation Results for {ticker}")
    print(company_results)

    # Showing yearly prediction results
    yearly_folder = "plots/yearly"

    print("\nShowing yearly prediction plots...\n")

    for file in os.listdir(yearly_folder):
        if file.startswith(ticker):
            img = mpimg.imread(os.path.join(yearly_folder, file))

            plt.figure(figsize=(8,4))
            plt.imshow(img)
            plt.axis("off")
            plt.title(file.replace(".png",""))
            plt.show()

    # Showing historical prices plot
    historical_path = f"plots/historical/{ticker}_history.png"

    if os.path.exists(historical_path):
        img = mpimg.imread(historical_path)

        plt.figure(figsize=(10,5))
        plt.imshow(img)
        plt.axis("off")
        plt.title(f"{ticker} Historical Forecast")
        plt.show()

    # Showing future forecasts
    future_path = f"plots/future/{ticker}_future.png"

    if os.path.exists(future_path):
        img = mpimg.imread(future_path)

        plt.figure(figsize=(10,5))
        plt.imshow(img)
        plt.axis("off")
        plt.title(f"{ticker} Future Forecast")
        plt.show()
        
```

```python
show_company_results("AAPL")
```
